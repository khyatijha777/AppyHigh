# Logistic regression

## What is logistic regression?
Logistic regression sounds to be a regression algorithm but it is actually a classification algorithm.
A classification algorithm is the algorithm which takes input data and predicts which class it belongs to.
i.e. we want to make a program that can classify an unknown fruit into apple or orange. Here the apple and orange are two classes. 

## Why can't we use linear regression for such problems?
Suppose we have problem like following, in such a problem one can use linear regression.
We can find the appropriate hypothesis function and for the value more than 0.5 , we can classify the nkonown fruit as apple and for less than 0.5 we can classify the fruit as orange.

But this is an ideal case. But when we work on real life problems then the data is not so simple to plot also this function can yield value which is more than 1 or negative. For real life problems this approach may lead to very wrong predictions as shown in the following figure-


## Why  this algorithm is called as regression?
Ths algorithm is called regression because it uses the same approach as linear or polynomial regression but uses a diffrent hypothesis function-
#### h(x) = 1/(1+e^(-z))
#### where z = theta(1) + x* theta(2)
#### or z = theta' * x
this hypothesis function also results in between 0-1.

if we get 0.7 then it means the probality that y = 1 is 70%
And probability that y = 0 is 30%



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


dataset = np.array([
    ["weight","diameter","class_of_fruit"],
    [180,7, 0],
    [120,5, 1],
    [100, 3, 2],
    [160,7, 0],
    [140,5, 1],
    [110, 3, 2],
    [189,7, 0],
    [129,5, 1],
    [109, 3, 2],
    [185,7, 0],
    [125,5, 1],
    [105, 3, 2],
])

# preparing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    dataset[1:, 0:2].astype(np.float), 
    dataset[1:, 2].astype(np.float), 
    test_size=.25,
)

# logistic regression model 

from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression(solver='lbfgs')

# training model
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
# predicting values for test data
print("test data :\n", x_test)
print("predicted class: ", y_pred)

for i in y_pred:
    if i==0:
        print("apple")
    elif i == 1:
        print("orange")
    elif i ==2:
        print("lemon")
```
