# Random Forest algorithm


## What is Random forest?
 
Random forest comes under supervised learning. It is a classification an algorithm which classifies data into several classes.But it can also be used for regression problems.

Random forest as the name suggests is a group of trees, In case of machine learning these trees are decision trees. This algorithm works on the "bagging" principle which says combination of results produced by various model will be more accurate. So the random forest predicts the result which is predicted by the majority of the decision trees.

Decision trees represents every single decision that can be made in a heirarichal order. And predicts the results based on its leaf nodes.

i.e. Random forest can classify a fruit into apple and orange. Here orange and apple are two classes.
So random forest is an algoithm which predicts the result based on decision which is taken by most of the trees.

## Why random forest?
There is no over fitting in random forest algorithm.

Training time is less.

Random forest shows high accuracy on large database. This algorithm shows better results with large databases.

It is able to handle both continuous and categorical variables.

This algorithm indication of which fields are most important for prediction or classification.

## Where is it used?
Random forest algorithem is used in multi object detection.

Virtual games are also a good example  of random forest algorithm where the movement of player is reflected into the the game. In such games movement and body parts of the player are indetified and accordingly the actions are taken in the game.

```python
# Decision Tree Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = [
    ["weight","diameter","class_of_fruit"],
    [180,7, 0],
    [120,5, 1],
    [100, 3, 2],
    [160,7, 0],
    [140,5, 1],
    [110, 3, 2],
    [189,7, 0],
    [129,5, 1],
    [109, 3, 2],
    [185,7, 0],
    [125,5, 1],
    [105, 3, 2],
]
dataset = np.array(dataset)
X = dataset[1:, 0:2]
y = dataset[1:, 2].T

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)

x_test = [[126,5]]
# Predicting a new result
y_pred = regressor.predict(x_test)
print("weight of the fruit: ", x_test[0][0])
print("radius of the fruit: ", x_test[0][1])
if y_pred ==0 :
    print("apple")
elif y_pred ==1 :
    print("lemon")
elif y ==2:
    print("orange")

```
