# Decision Tree

## What is Decision Tree?
A decision tree is graphical representaion of all the possible decisions that one can take based on certain condition. It helps us visualize why any particular decision was taken.


## Why it is called a tree?
This is called as decision tree because it starts with a single decision than keeps taking decision which can be represented as branches of a tree.
i.e. In the above diagram at the root, the decision is taken than any user wants to buy fruit or juice, if he is hungry then the decision is handed over to the left tree and if thirsty than the decision is handed over to the right tree.
lets say the user is hungry, then the decisiion is taken that if the user like mango  or oranges and the final decision is made based on previous decision.
## What is root?
ROoot is the tope most node of the tree where the whole data is feeded in to the tree. It resembels the whole set of inputs. Which is further devided by the branches of the tree.
## How the branches are decided?
The branches of the tree consist of various boolean type question. These questions are based on our data.
The process of separating data into various branches is called splitting.
And when no further splitting is possible, then the node is defined as leaf node.


```python
# Decision Tree Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = [
    ["weight","diameter","class_of_fruit"],
    [180,7, 0],
    [120,5, 1],
    [100, 3, 2],
    [160,7, 0],
    [140,5, 1],
    [110, 3, 2],
    [189,7, 0],
    [129,5, 1],
    [109, 3, 2],
    [185,7, 0],
    [125,5, 1],
    [105, 3, 2],
]
dataset = np.array(dataset)
X = dataset[1:, 0:2]
y = dataset[1:, 2].T

print(y)
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)

X_test = [[100,3]]
# Predicting a new result
y_pred = regressor.predict(X_test)
if y_pred == 0:
    print("apple")
elif y == 1:
    print("orange")
if y_pred == 2:
    print("lemon")
```
