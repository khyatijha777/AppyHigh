#### get_K_Neighbors function takes all the trainig data, test data and no of neighbors to find 
#### Calculates distance of all training data instances with test data instance using euclidean function
#### Finds K nearest instances from training data , and returns them as list called neighbors

```python
import operator 
def get_K_Neighbors(train_data, test_data, k):
    distances = []
    length = len(test_data)-1
    for x in range(len(train_data)):
        dist = euclidean(test_data, train_data[x], length)
        distances.append((train_data[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```
